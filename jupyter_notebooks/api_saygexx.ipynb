```python
import pandas as pd
import requests
from datetime import date, datetime

load_date = date.today() # Дата загрузки данных
response = requests.get('https://api.open-meteo.com/v1/forecast', # Запрос к API  с данными о погоде, задаем параметры.
params = {
	"latitude": 58.0105,
	"longitude": 56.2502,
	"hourly": ["temperature_2m", "snowfall"],
	"timezone": "auto",
	"start_date": load_date ,
	"end_date": load_date ,
}
)
response.raise_for_status() # Проверяем на ошибки, ответ от сервиса с погодой

data = response.json() # Один из вариантов представления данных в формате JSON
data["updated_at"] = pd.to_datetime(datetime.now()) # вставляем дату загрузки данных в таблицу
df = pd.json_normalize(data) # Подставляем в метод пандас нащи данные в формате JSON

df.head() # Метод в библиотеке pandas который показывает первые несколько строк в таблице

file_name = f'Perm_{load_date}.csv' #Cохраняем в csv

if response: # Если диапазон от 200 до 300, то ошибки нет и ответ пришел и файл должен сохраниться 
	df.to_csv(file_name,index = False) # Если в таблице есть строки то сохраняем, False убирает лишний индекс из таблицы
```
