```python
import requests
import pandas as pd
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
response = requests.get(url,
     params={
        "format": "geojson",
        "latitude": 35.42,
        "longitude": 139.46,
        "maxradiuskm": 100,
        "starttime": "2025-01-01",
        "endtime": "2026-01-01",
        "limit": 100
        } 
    )

response.raise_for_status()
data = response.json()

df = pd.json_normalize(data["features"], sep='_') 
df = df[[
    "properties_mag",
    "properties_place",
    "properties_time",  # бизнес дата
    "geometry_coordinates"
]]
df["properties_time"] = pd.to_datetime(df["properties_time"], unit="ms") 
df["updated_at"] = pd.Timestamp.utcnow() # Дата выгрузки данных 
csv_path = "tmp//Tokyo_earthquake.csv"
if len(df) > 0:
    df.to_csv(csv_path, index=False)
```
