# GitHUb API
## List public events

```python
import pyarrow
import pandas as pd
import requests
import logging
from datetime import datetime

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(name)s | %(message)s"
)
logger = logging.getLogger(__name__)

csv_path = f'/home/coder/tmp/github_list_pub_events.csv'
```

```python
headers = {
    "Accept": "application/vnd.github+json"
}

params = {
    "per_page" : 100,
    "page" : 3
}

response = requests.get('https://api.github.com/events', headers= headers , params=params)

update_dt = datetime.now().isoformat()

data = response.json()

df = pd.json_normalize(data)

logger.info(df.info())
```

```python
# Столбцы без единого значения NaN
df_no_nan = df.dropna(axis=1, how='any')

#Добавление столбца в df строки обновления
df_no_nan.insert(0, 'update_at', datetime.now())

logger.info(df_no_nan.info())
```

```python
df_no_nan.to_csv(csv_path, index=False)
```
