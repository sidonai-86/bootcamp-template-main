```python
import requests
from datetime import datetime
import pandas as pd

load_date = datetime.now().strftime("%Y-%m-%d")

url = "https://earthquake.usgs.gov/fdsnws/event/1/query"


response = requests.get(
    url,
    params={
        "format": "geojson",
        "endtime": "2026-01-15",
        "starttime": "2026-01-01",
        "minmagnitude": "6",
    },
)

response.raise_for_status()

data = response.json()

# print(data)

# # 1. Извлекаем данные
# quakes = [feature['properties'] for feature in data['features']]
# df = pd.DataFrame(quakes)

# # 2. Преобразуем время
# df['datetime'] = pd.to_datetime(df['time'], unit='ms')

# # 3. Форматируем: оставляем только дату и часы:минуты
# df['date'] = df['datetime'].dt.strftime('%Y-%m-%d')
# df['time_only'] = df['datetime'].dt.strftime('%H:%M')

# # 4. Собираем итоговую таблицу
# result = df[['place', 'date', 'time_only', 'mag']]

# print(result.head())

# df.head()

df = pd.json_normalize(data)

# df.head()

csv_path = f"tmp/eq_{load_date}.csv"

df["update_at"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

df.to_csv(csv_path, index=False)
```
