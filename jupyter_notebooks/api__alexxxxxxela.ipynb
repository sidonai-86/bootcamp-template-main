```python
import requests
import json
import pandas as pd
from io import StringIO
from datetime import datetime, date, timedelta
```

```python
URL = "https://earthquake.usgs.gov/fdsnws/event/1/query"
datte = date(2026, 1, 1)
params = {
    "format": "csv",
    "starttime": datte.strftime("%Y-%m-%d"),
    "endtime": (datte + timedelta(days=1)).strftime("%Y-%m-%d") ,
}
result = requests.get(url=URL, params=params)
df = pd.read_csv(StringIO(result.text))
```

```python
df['update_at'] = pd.to_datetime(datetime.now())
# df.to_csv('/tmp/eq_data_month.csv')
```
