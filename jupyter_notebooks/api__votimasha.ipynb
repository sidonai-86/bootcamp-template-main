API Weather: https://open-meteo.com/en/docs

```python
import requests
import pandas as pd
import pyarrow
```

```python
url = 'https://api.open-meteo.com/v1/forecast'

# погода в Кижах, Карелия, за декабрь 2025 года
e_date = '2025-12-31'
s_date = '2025-12-01'
response = requests.get(url, params={"latitude": 62.08, "longitude": 35.21, "hourly": ['temperature_2m', 'apparent_temperature', 'rain', 'snowfall'], "timezone": 'auto', "start_date": s_date, "end_date": e_date})
```

```python
response.raise_for_status() # ошибку не дает

data = response.json()
print(data)

df = pd.json_normalize(data)
df.head()

#df.to_csv('csv_weather_kizhi.csv', index=False)
```

```python
# красивая таблица + дата выгрузки
new_df = pd.DataFrame({'date and time': data['hourly']['time'], 'real temperature': data['hourly']['temperature_2m'], 'temperature feels like': data['hourly']['apparent_temperature'], 'rain': data['hourly']['rain'], 'snow': data['hourly']['snowfall']})
new_df['upload date'] = pd.to_datetime("today").strftime("%Y-%m-%d")
new_df.head()
```

```python
# сохраняем в файлы
new_df.to_csv('csv_weather_kizhi.csv', index=False)
new_df.to_parquet('csv_weather_kizhi.parquet', index=False)
```
