```python
headers = {
    "Authorization": f"OAuth {ACCESS_TOKEN}"
}

metrics = [
    "ym:s:visits",
    "ym:s:users",
    "ym:s:newUsers",
    "ym:s:pageviews"
]

dimensions = [
    "ym:s:regionCountry",
    "ym:s:regionCity",
    "ym:s:date"
]

params = {
    "date1": "2026-01-01",
    "date2": "2026-01-18",
    "ids": COUNTER_ID,
    "metrics": ",".join(metrics),
    "dimensions": ",".join(dimensions)
}
```

```python
response = requests.get(url, headers=headers, params=params)
data = response.json()

dimension_nm = data["query"]["dimensions"]
metric_nm = data["query"]["metrics"]
```

```python
rows = []
for row in data['data']:
    dimension_values = [d['name'] for d in row['dimensions']]
    metrics = row['metrics']
    rows.append(dimension_values + metrics)

columns = dimension_nm + metric_nm

df = pd.DataFrame(rows, columns=columns)

df["update_at"] = datetime.now()
```
