```python
import requests
import pandas as pd
from datetime import datetime
```

```python
start_time = '2026-01-16'
end_time = '2026-01-17'

url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'

response = requests.get(url,
    params={
        'format': 'geojson',
        'starttime': start_time,
        'endtime': end_time
    }
)

data = response.json()
df = pd.json_normalize(data['features'])

df['business_date'] = pd.to_datetime(df['properties.time'], unit='ms')
df['update_at'] = datetime.now()

cvs_path = 'tmp/hv1_1'
df.to_csv(cvs_path, index=False)
```
