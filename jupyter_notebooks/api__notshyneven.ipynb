```python
import pandas as pd
import pyarrow 
import requests
from datetime import datetime, timezone



minmagnitude = 5                                                # допустим нужны данные о землетрясениях магнитудой >= 5     


url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'        # берем url

response = requests.get(url,                                    # запрос                                                            # минимальную магнитуду 5, период - в месяц
    params={'format':'geojson',                            
            'minmagnitude': 5,
            'starttime': '2025-12-15',
            'endtime': '2026-01-15'}
)

response.raise_for_status()                                     # проверяю чтобы не было ошибок

data = response.json()                                          # превращает данные в словарь

features = data['features']                                     # выбираю все события из словаря


df = pd.json_normalize(features)                                # преобразую в DataFrame

df["updated_at"] = datetime.now(timezone.utc).replace(microsecond=0)

csv_path = 'api.notshyneven.csv'                                # переменная сsv

df.to_csv(csv_path, index = False)                              # создаю csv - файл  
```
