```python
## Создание таблиц
""" 
Создание таблиц в ClikHouse очень похоже на создание таблиц в других базах данных, за исключением того что 
в КХ обязательно нужно указывать движок базы данных. Каждый движок во своему уникален, но самый встречаемый 
и вообще именно для него и создавался КХ это MergeTree.
"""

client.command(f'''
    DROP TABLE IF EXISTS {database}.mt;
''')

client.command(f'''
    CREATE TABLE IF NOT EXISTS {database}.mt
    (
        id UInt32,
        dt datetime
    )
    ENGINE = MergeTree                                                 -- обязательно нужно указывать движок
    PRIMARY KEY(id)                                                    -- не обязательное поле по умолчанию равно ORDER BY
    ORDER BY(id, dt)                                                   -- обязательно должны быть колонки в порядке из primary key
    TTL dt + INTERVAL 1 MONTH DELETE;                                  -- от времени dt отсчитывается 1 месяц после чего данные удаляются (данные удаляются полсе слияния)
        --dt + INTERVAL 1 MONTH DELETE WHERE toDayOfWeek(d) IN (6, 7); -- удаление с фильтрацией
        --dt + INTERVAL 1 WEEK TO VOLUME 'aaa',                        -- перемещение данных в вольюм(совокупность дисков, задаваемая в конфигах)
        --dt + INTERVAL 2 WEEK TO DISK 'bbb';                          -- перемещение данных на диск (указывается имя диска)
''')

```

```python
# Самостоятельная работа на проверку работоспостобности TTL:
"""
TTL в ClickHouse — это механизм управления временем жизни данных в таблицах на движке MergeTree, 
позволяющий автоматически удалять, перемещать или агрегировать устаревшие строки/столбцы.
Для столбцов: MODIFY COLUMN balance TTL time + INTERVAL 2 HOUR — заменяет значения дефолтными.
Условия: TTL time + INTERVAL 1 MONTH DELETE WHERE event = 'error' — выборочное удаление.
События TTL активируются только во время слияний (merge) частей данных, не мгновенно.
- По умолчанию раз в 4 часа (merge_with_ttl_timeout = 14400 сек).
- При горячем/холодном хранении: TTL time + INTERVAL 1 MONTH TO VOLUME 'cold' — перемещение на другой диск.
Особенности
- Партиционирование по дате ускоряет удаление целых партиций (ttl_only_drop_parts=1).
- Поддержка интервалов: SECOND, HOUR, DAY, MONTH, YEAR.
- Если слияний нет — данные остаются до ближайшего merge.
"""

#1) Создай таблицу с движком MergeTree c 2мя колонками: id Int32, dt DateTime
#2) Укажи только сотировку по id. PRIMARY KEY указывать не нужно оно будет совпадать по умолчанию с ORDER BY
#3) Напиши в условии TTL удаление строки по dt через 10 секунд
client.command(f'''
    DROP TABLE IF EXISTS {database}.test_ttl;
''')

client.command(f'''
    CREATE TABLE IF NOT EXISTS {database}.ttest_ttl
    (
        id UInt32,
        dt datetime
    )
    ENGINE = MergeTree                                                 -- обязательно нужно указывать движок
    ORDER BY(id)                                                       -- обязательно должны быть колонки в порядке из primary key
    TTL dt + INTERVAL 10 second DELETE;                                  -- от времени dt отсчитывается 1 месяц после чего данные удаляются (данные удаляются полсе слияния)
''')
```

```python
# Самостоятельная работа на расширенные атрибуты колонок:
"""
Расширенные атрибуты колонок в ClickHouse — это дополнительные свойства, 
задаваемые для столбцов в таблицах MergeTree при создании или через ALTER, помимо типа данных.
Основные атрибуты
CODEC: Сжатие данных (например, CODEC(LZ4) для быстрого сжатия или DoubleDelta для временных рядов).
TTL: Время жизни столбца (удаление/перемещение, как TTL time + INTERVAL 1 MONTH).
DEFAULT/MATERIALIZED/ALIAS: Выражения по умолчанию (DEFAULT now(), MATERIALIZED today() вычисляется всегда).
SETTINGS: Параметры хранения (max_compress_block_size=65536 для размера блока сжатия).
"""
"""
1) Создайте таблицу `product_sales`, которая будет хранить информацию о продажах товаров. 
В таблице необходимо использовать **все указанные выше атрибуты хотя бы по одному разу**.

| Поле          | Тип данных  | Атрибуты                                                              |
|---------------|-------------|------------------------------------------------------------------------|
| `sale_id`     | `UInt64`    | `COMMENT`: "Уникальный идентификатор продажи"                        |
| `price`       | `Float32`   | `DEFAULT`: 0.0                                                        |
| `quantity`    | `UInt8`     | `DEFAULT`: 1                                                          |
| `total`       | `Float32`   | `MATERIALIZED`: `price * quantity`                                   |
| `description` | `String`    | `CODEC`: `ZSTD(1)`                                                    |
| `taxed_total` | `Float32`   | `ALIAS`: `total * 1.2`

"""
client.command(f'''
    DROP TABLE IF EXISTS {database}.product_sales;
''')


client.command(f'''
    CREATE TABLE {database}.product_sales
    (
        sale_id UInt64 COMMENT 'Уникальный идентификатор продажи',
        price Float32 DEFAULT 0.0,
        quantity UInt8 DEFAULT 1,
        total Float32 MATERIALIZED price * quantity,
        description String CODEC(ZSTD(1)),
        taxed_total Float32 ALIAS total * 1.2
    )
    ENGINE = MergeTree
    ORDER BY sale_id;
''')

```

```python
## Партицирование
"""
Партицирование - это процессобъединения наборов данных по единому критерию, например по месяцу. 
Это позволяет ускорить процес считывания данных по выбранному критерию при фильтрации. 
В ClickHouse существует 4 типа партицирования
В ClickHouse партиции в таблицах MergeTree создаются автоматически на основе выражения PARTITION BY, 
а не вручную — это ключевое отличие от традиционных БД.

Способы задания ключа партиционирования
Способ	Синтаксис	Пример	Особенности
По столбцу	PARTITION BY (column)	
    PARTITION BY toYYYYMM(date)	Партиции по месяцам (202601, 202602). Стандарт для временных рядов.
Пользовательский	PARTITION BY (expr1, expr2)	
    PARTITION BY (toMonday(date), region_id)	Кортеж выражений. Меньше партиций, но сложнее управление.
Без партиций	Нет PARTITION BY	
    ENGINE = MergeTree ORDER BY id	Одна большая партиция. Подходит для <10M строк.

Как это работает
При INSERT ClickHouse хэширует значение ключа партиции и создаёт отдельный кусок данных для каждого уникального значения.
Партиции видны в system.parts (столбец partition).
Удаление: ALTER TABLE DROP PARTITION '202601' или TTL с ttl_only_drop_parts=1 для целых партиций.

Сравнение подходов
text
-- Много мелких партиций (по дням)
PARTITION BY toDate(event_time)  -- 365 партиций/год

-- Меньше партиций (по месяцам)  
PARTITION BY toYYYYMM(event_time)  -- 12 партиций/год

-- Комплексный ключ
PARTITION BY (toYYYY(event_time), user_type)  -- 2×12=24 партиции/год

Рекомендация: 100-1000 партиций оптимально. Слишком много (>10k) замедляет DROP/SELECT

Способы PARTITION BY expr в таблицах MergeTree:
Диапазоном  PARTITION BY age_group где age_group = (age / 10) или числовые диапазоны.
    Партиции по диапазонам значений (0-9, 10-19 лет).
    Когда: Демография, категории, неравномерные данные.
    PARTITION BY (value / 1000)  -- партиции 0,1,2,... по тысячам
    PARTITION BY region_tier  -- S1,S2,A1,A2 (иерархия)
Интервалом (по времени) PARTITION BY toYYYYMM(date), toStartOfDay(event_time), toMonday(date).
    Партиции = временные интервалы (месяц=202601, день=2026-02-04).
    Когда: Временные ряды, логи — стандарт для большинства таблиц.
    Плюсы: Логично DROP старых (DROP PARTITION '202501'), быстрое сканирование.
    PARTITION BY toYYYYMM(ts)  -- 12 партиций/год
    PARTITION BY toStartOfWeek(ts)  -- 52 партиции/год
Хеш-функцией    PARTITION BY intHash32(city_id) % 100 или sipHash64(user_id) % 16.
    Данные распределяются равномерно по фиксированному числу партиций (например, 100).
    Когда: Для равномерной нагрузки при GROUP BY или когда нет естественного порядка (user_id, session_id).
    Плюсы: Предсказуемо, нет "горячих" партиций.
    Минусы: Невозможно DROP по смыслу (непонятно, что удалять).
    PARTITION BY cityHash64(region, shop_id) % 50  -- 50 равных партиций
Комбинированный Кортеж выражений PARTITION BY (toYYYYMM(date), region).
    Комбинация времени + категория (202601_SPB, 202601_MSK).
    Когда: Многоуровневое управление (DROP по региону + времени).
    PARTITION BY (toYYYY(event_time), user_type)  -- 12×2 = 24 партиции/год
    PARTITION BY (toMonday(date), intHash32(dept_id) % 10)  -- время + хеш

Сравнение эффективности
Способ	        #Партиций	            DROP удобен	    Селективность WHERE	    Пример
Хеш	            Фиксировано (10-1000)	❌ Нет	        ❌ Равномерно	       % 100
Интервал	    Растёт со временем	    ✅ Идеально	    ✅ Высокая	           toYYYYMM()
Диапазон	    Стабильно	            ✅ Логично	    ✅ Средняя	           rand() % 10
Комбинированный	Много	                ✅ Гибко	        ✅ Максимальная	       (month, region)

Оптимум: 100-1000 активных партиций. Используйте system.parts для мониторинга.
"""
```

```python
## Движки таблиц
"""
Движки таблиц в ClickHouse определяют, как хранятся и обрабатываются данные — от простых логов до распределённых систем.

Семейство MergeTree (основное для продакшена)
Универсальные движки для больших объёмов с колоночным хранением, индексацией и фоновыми merge.

MergeTree: Базовый, с партициями и ORDER BY. Для временных рядов.
ReplacingMergeTree: Удаляет дубликаты по ключу при merge (версия опциональна).
SummingMergeTree: Суммирует числовые столбцы для строк с одинаковым ключом.
AggregatingMergeTree: Хранит агрегированные состояния (-State типы).
CollapsingMergeTree: Поддержка отмены записей через sign колонку.
Replicated* (ReplicatedMergeTree и др.): +репликация через ZooKeeper.

Семейство Log (малые объёмы)
Для временных таблиц без индексов, данные в N файлов.

TinyLog: Несжатые данные (~1GB макс).
Log: Сжатие LZ4.
StripeLog: Полосы данных для стриминга.

Интеграции и специальные

Движок	                Назначение	                        Хранение данных
Distributed	            Распределённые запросы по шардам	Нет, прокси
Buffer	                Буферизация INSERT в MergeTree	    RAM → диск
Memory	                Полностью в RAM, без сжатия	        RAM
Dictionary	            Словари для IN/JOIN	                RAM/файлы
Kafka	                Стриминг из Kafka	                Буфер
File/URL	            Файлы/OS, HTTP источники	        Файлы
Null/GenerateRandom	    Тесты, пустые таблицы	            Нет
View/MaterializedView	Представления, автоматизация	    SELECT/INSERT

Выбор движка

-- Временные ряды: MergeTree PARTITION BY toYYYYMM(ts) ORDER BY (user_id, ts)
-- Дедупликация: ReplacingMergeTree ORDER BY (id, ver)
-- Агрегация: SummingMergeTree ORDER BY key
-- Кластер: ReplicatedMergeTree ORDER BY id
MergeTree покрывает 90% случаев. Используйте system.tables для анализа.
"""
```

```python
## Агрегаторные типы данных
"""
Сущесвует 2 агригаторных типа данных:

    SimpleAggregateFunction - предназначет для хранения простых агрегатов, которое хранит конечно состояние
    AggregateFunction - сложные агрегаты, которая хранит состояние всех добавленных значений

Разберем пример с SimpleAggregateFunction
"""
client.command(f'''
    DROP TABLE IF EXISTS {database}.simple;
''')

client.command(f'''
    CREATE TABLE {database}.simple (
      id UInt64, 
      val_sum SimpleAggregateFunction(sum, UInt64), -- предусмотрен для хранения простых агрегатов(хранит конечное состояние)
      val_max SimpleAggregateFunction(max, UInt32)
    ) 
    ENGINE=AggregatingMergeTree 
    ORDER BY id;
''')

client.command(f'''
    INSERT INTO {database}.simple SELECT  1, number, number from numbers(10);
''')
client.command(f'''
    INSERT INTO {database}.simple SELECT  2, sum(number), max(number) from numbers(5);
''')
client.command(f'''
    INSERT INTO {database}.simple SELECT  1, number, number from numbers(8);
''')
```

```python

"""
Разберем пример с AggregateFunction

Здесь немного все посложнее так как нужно использовать специальные комбинаторы
Комбинаторы агрегаторных типов данных:

    SimpleState — возвращает результат агрегирующей функции типа SimpleAggregateFunction.
    State — возвращает промежуточное состояние типа AggregateFunction, используется при вставке.
    Merge — берёт множество состояний, объединяет их и возвращает результат полной агрегации данных.
    MergeState — выполняет слияние промежуточных состояний агрегации, возвращает промежуточное состояние агрегации.
"""
client.command(f'''
    DROP TABLE IF EXISTS {database}.aggr_func_tbl;
''')

client.command(f'''
    CREATE TABLE {database}.aggr_func_tbl
    (
        id UInt64,
        val_uniq AggregateFunction(uniq, UInt64),         -- Хранит в себе промежуточное состояние данных
        val_any AggregateFunction(anyIf, String, UInt8),
        val_quant AggregateFunction(quantiles(0.5, 0.9), UInt64)
    ) ENGINE=AggregatingMergeTree 
    ORDER BY id;
''')

client.command(f'''
    INSERT INTO {database}.aggr_func_tbl
    SELECT 
        1, 
        uniqState(toUInt64(rnd)),                 -- кол-во уникальных значений
        anyIfState(toString(rnd),rnd%2=0),
        quantilesState(0.5, 0.9)(toUInt64(rnd)) 
    FROM
        (SELECT arrayJoin(arrayMap(i -> i * 10, range(10))) as rnd);
''')
```

```python

## Отправляем данные в кафку
"""
Apache Kafka — это распределённая платформа для потоковой обработки данных в реальном времени, 
работающая как высокопроизводительный брокер сообщений.

Основная архитектура
    Producer (производитель): отправляет сообщения в топики.
    Topic (тема): лог сообщений, разбитый на партиции для параллелизма.
    Broker (сервер): хранит и реплицирует данные в кластере.
    Consumer (потребитель): читает из топиков, отслеживая offset (смещение).

Ключевые особенности
    Долговременное хранение: сообщения не удаляются мгновенно, а хранятся по retention policy (дни/недели).
    Репликация: каждая партиция копируется на несколько брокеров (replication factor).
    Порядок: строгий порядок внутри партиции, но не между ними.
    Масштабирование: горизонтальное — добавляйте брокеры/ZooKeeper/KRaft.

Применение в data engineering%
    Логи → Kafka → ClickHouse (ваш стек)
        logs → Kafka(topic=events) → Consumer → INSERT INTO clickhouse_table
    Потоковая аналитика (Kafka Streams, ksqlDB).
    Связь микросервисов (замена RabbitMQ для high-throughput).
    Data pipelines (ETL в реальном времени).

Сравнение с очередями
Kafka	                    RabbitMQ	        Redis
Диски, триллионы сообщений	RAM+диск, pub/sub	RAM, кэш
Потоки событий	            Задачи очередей	    Кэш/pub-sub
Дни хранения	            Секунды	            Секунды

В вашем стеке: Kafka + ClickHouse = идеальная связка для логов/метрик (Kafka как буфер, CH как хранилище).
"""

client.command(f'''
    DROP TABLE IF EXISTS {database}.kafka_out_message;
''')

client.command(f'''
    CREATE TABLE {database}.kafka_out_message
    (
        id UInt32
    )
    ENGINE = Kafka
    SETTINGS
        kafka_broker_list = 'kafka:29093',
        kafka_topic_list = '{database}.clickhouse.topic', -- создай такой же топик
        kafka_group_name = 'clickhouse_consumer_group',
        kafka_format = 'JSONEachRow';
''')   

# <clickhouse_connect.driver.summary.QuerySummary at 0x7e875b405250>
```

```python

## Самостоятельная работа на движок Kafka
"""
Ты уже много чего прошел, пришло время давать серьезные задачи, как из реально дают в компаниях.(Это реальный кейс с работы)

К тебе приходит тимлид и говорит: Аналитики из команды продаж, 
просят предоставить им данные от внутреннего сервиса отвечающего за продажи определенных вещей. 
Они уже договорили с DBA(администратор баз данных) отвечающего за поддержание БД данного сервиса и 
он настроил топик в кролике(Rabbitmq, тоже самое что и кафка только в профиль), 
который находится в их структуре. 
Тебе необходимо положить эти данные в GreenPlum. 
Ты знаешь, что у вас в команде есть ClikHouse, Kafka, Spark, GreenPlum, 
но из Spark нельзя достучаться до кролика, так как он находится в другой сети. 
Spark может читать стримовые данные только из локальной кафки. 
Доступ есть у Клика, так как та самая комадна кладёт туда свои данные для анализа.

Твоя задача: взять данные из кролика(в роли кролика у нас выступает кафка с топиком source.public.dbz_heartbeat) 
и загнать данные в другой топик с именем <фамилия>.source.data_from_rabit. 
От туда твой мнимый товарисчь, мнимо данные заберёт и положит в мнимый GreenPlum
"""
"""
Логика задачи (ETL: RabbitMQ → GreenPlum)
Цель: Перенести стриминговые данные из изолированного RabbitMQ (здесь Kafka-симулятор: source.public.dbz_heartbeat) 
в GreenPlum через вашу локальную Kafka. Почему: Spark не видит RabbitMQ (разные сети), но ClickHouse/Kafka доступны.

Архитектура
RabbitMQ (topik dbz_heartbeat) 
    ↓ (connector/consumer)
ClickHouse (Kafka Engine) ← читает
    ↓ (Spark Streaming)
Локальная Kafka (ваш.topik.data_from_rabit) 
    ↓ (товарищ)
GreenPlum

ClickHouse как bridge (доступен везде) → обходит сетевые ограничения RabbitMQ.
Spark только локальная Kafka → используем CH как прокси.
Материализация + TTL → не перегружаем, чистим старое.

Аналогично S3→CH (cite:119), 
Airflow для оркестрации (cite:121).

Масштаб: Batch → stream, checkpointing
"""
```

```python
 

# Решение(Смотреть только после выполенения)
# В ClickHouse таблицу для чтения топика
client.command(f'''
    DROP TABLE IF EXISTS {database}.data_from_rabbit;
''')

client.command(f'''
   create table {database}.data_from_rabbit 
      (
         json_kafka String
      )
   Engine = Kafka
   SETTINGS
        kafka_broker_list = 'kafka:29093',
        kafka_topic_list = 'source.public.dbz_heartbeat',
        kafka_group_name = 'clickhouse_consumer_group',
        kafka_format = 'JSONAsString'; 
''')

client.command(f'''
    DROP TABLE IF EXISTS {database}.data_rabbit_to_kafka;
''')

client.command(f'''
   create table {database}.data_rabbit_to_kafka 
   (
      before String,
      after String
   )
   Engine = Kafka
   SETTINGS
        kafka_broker_list = 'kafka:29093',
        kafka_topic_list = 'shushiato.source.data_from_rabit',
        kafka_group_name = 'clickhouse_consumer_group',
        kafka_format = 'JSONEachRow';
''')
```

```python

# Создать промежуточную таблицу (материализация)
client.command(f'''
    DROP VIEW IF EXISTS {database}.data_from_rabbit_mv;
''')

client.command(f'''
   CREATE MATERIALIZED VIEW {database}.data_from_rabbit_mv TO {database}.data_rabbit_to_kafka AS 
      SELECT 
         simpleJSONExtractRaw(json_kafka, 'before') as before,
         simpleJSONExtractRaw(json_kafka, 'after') as after
         -- для примера тут приведено всего 2 строки, но надеюсь принцип вы поняли
      from {database}.data_from_rabbit ;
''')
```

```python
## Реши задачу с собесендования
# Решение(Смотреть только после выполенения)
"""
Дано: Представим, что у нас есть интернет магазин, из которого мы собираем данные.

    данные по логинам пользователей

    данные летят в нашу таблицу в режиме реального времени, каждый раз, когда пользователь логинится на сайт
    таблица: user_login
    поля:
    user_id Uint64
    login_time DateTime
    объем данных большой

2.данные по акциям

    данные летят в нашу таблицу в режиме реального времени, каждый раз, когда создается новая акция
    таблица event
    поля:
    event_name String
    start_time DateTime
    end_time DateTime
    объем данных маленький

Нужно посчитать кол-во логинов пользователя во время действия акйций за выбранный промежуток времени
"""
```

```python
# user_login
client.command(f'''
    CREATE TABLE IF NOT EXISTS {database}.user_login
    (
        user_id UInt64,
        login_time DateTime
    )
    ENGINE = ReplacingMergeTree
    PARTITION BY toYYYYMM(login_time)     -- Месячные партиции
    ORDER BY (login_time, user_id)        -- Primary key для time-range запросов
    TTL toDateTime(login_time) + INTERVAL 1 YEAR  -- Автоудаление >1 года
    SETTINGS 
    index_granularity = 8192,
    merge_with_ttl_timeout = 14400;  -- Быстрые TTL merges
''')

```

```python
# заполним тестовыми данными
client.command(f'''
    INSERT INTO shushiato.user_login 
    SELECT 
        number % 10000 + 1 AS user_id,
        addSeconds(toDateTime('2026-02-01 00:00:00'), number % 2592000) AS login_time  -- 30 дней (2.592M сек)
    FROM system.numbers 
    LIMIT 100000;
''')
```

```python
# проверка
client.command(f'''
    SELECT user_id, login_time, count(*) as login_count FROM shushiato.user_login
        where login_time between '2026-02-01' and '2026-02-14'
        group by user_id, login_time;
''')
```

```python
# event
client.command(f'''
    CREATE TABLE IF NOT EXISTS shushiato.event (
        event_name String,
        start_time DateTime,
        end_time DateTime
    )
    ENGINE = ReplacingMergeTree 
    PARTITION BY toYYYYMM(start_time)    
    ORDER BY (start_time, event_name)  -- Для BETWEEN запросов
    TTL end_time + INTERVAL 2 YEAR   -- Акции храним 2 года
    SETTINGS 
        index_granularity = 4096;  -- Меньше для small data
''')
```

```python
# строка для каждой даты в акции
client.command(f'''
    CREATE TEMPORARY TABLE temp_events
        (
        event_name String,
        event_time Date
    );
''')
```

```python

client.command(f'''
    INSERT INTO temp_events
    SELECT 
        event_name,
        toDate(addDays(toDate(start_time), n)) AS event_time
    FROM shushiato.event
    ARRAY JOIN range(toUInt32(dateDiff('day', start_time, end_time) + 1)) AS n
    ORDER BY event_name, event_time;
''')
```

```python
# Нужно посчитать кол-во логинов пользователя во время действия акйций за выбранный промежуток времени
client.command(f'''
    SELECT ul.user_id, te.event_name, count(*) as login_count 
        FROM temp_events as te
    JOIN shushiato.user_login as ul
        --логин в промежутке акции
    ON toDate(ul.login_time) = te.event_time
    --логин в заданном интервале
    WHERE ul.login_time BETWEEN '2026-01-01' AND '2026-03-12'
    --and ul.user_id = 7019
    GROUP BY ul.user_id, te.event_name
    ORDER BY ul.user_id, te.event_name;
''')
```
