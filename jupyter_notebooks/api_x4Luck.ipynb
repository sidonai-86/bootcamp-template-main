```python
base_url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
output_dir = '/home/coder/temp/'
```

```python
#current_date  = datetime.now()
#too_weaks_ago = current_date - timedelta(days=14)
params = {
    'format':'geojson',
    'starttime':too_weaks_ago.strftime('%Y-%m-%d'),
    'endtime':current_date.strftime('%Y-%m-%d'),
}
response  = requests.get(base_url, params=params)
```

```python
def convert_unix_date(date:int) -> datetime:
    return datetime.fromtimestamp(date/1000)
```

```python
data = [features.get('properties') for features in response.json().get('features')]
df = pd.DataFrame(data)
df['business_date'] = df['time'].apply(convert_unix_date) # Переводим время землетресения из UNIXtime
df['update_at'] = current_date
```

```python
def file_name_generate(start_date:datetime, end_date:datetime, format:str) -> str:
    start_date = start_date.strftime('%Y_%m_%d')
    end_date = end_date.strftime('%Y_%m_%d')
    return f'earthquake_{start_date}_{end_date}.{format}'
```

```python

file_name = file_name_generate(too_weaks_ago, current_date, 'csv')
output_path = os.path.join(output_dir, file_name)
df.to_csv(output_path, sep='~')
```
