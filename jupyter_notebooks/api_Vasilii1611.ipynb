```python
import requests
import pandas as pd
from datetime import datetime, timedelta
```

```python
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
    
# Параметры для последней недели
end_date = datetime.now()
start_date = end_date - timedelta(days=7)
    
params = {
        'format': 'geojson',
        'starttime': start_date.strftime('%Y-%m-%d'),
        'endtime': end_date.strftime('%Y-%m-%d'),
        'minmagnitude': 1.0,
        'orderby': 'time',
        'limit': 1000
    }
```

```python
response = requests.get(url, params=params)
data = response.json()
```

```python
earthquakes = []

for feature in data['features']:
    props = feature['properties']
    geometry = feature['geometry']
            
    earthquake = {
        'time': pd.to_datetime(props['time'], unit='ms'),
        'latitude': geometry['coordinates'][1],
        'longitude': geometry['coordinates'][0],
        'depth_km': geometry['coordinates'][2],
        'magnitude': props['mag'],
        'place': props['place'],
        'type': props['type'],
        'status': props['status'],
        'tsunami': props['tsunami'],
        'significance': props['sig'],
        'url': props['url'],
        'load_dt': datetime.now().strftime('%Y-%m-%d')
        }

    earthquakes.append(earthquake)
```

```python
#Создаем пандовский DF
df = pd.DataFrame(earthquakes)

#Записываем в csv
df.to_csv('Earthquakes.csv')
```

```python
#Функция на всякий случай для дальнейшей работы))
def get_earthquakes_dataframe():
    """
    Получение данных и преобразование в DataFrame
    """
    url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
    
    # Параметры для последней недели
    end_date = datetime.now()
    start_date = end_date - timedelta(days=7)
    
    params = {
        'format': 'geojson',
        'starttime': start_date.strftime('%Y-%m-%d'),
        'endtime': end_date.strftime('%Y-%m-%d'),
        'minmagnitude': 1.0,
        'orderby': 'time',
        'limit': 1000
    }
    
    try:
        response = requests.get(url, params=params)
        data = response.json()
        
        # Извлекаем данные из GeoJSON
        earthquakes = []
        for feature in data['features']:
            props = feature['properties']
            geometry = feature['geometry']
            
            earthquake = {
                'time': pd.to_datetime(props['time'], unit='ms'),
                'latitude': geometry['coordinates'][1],
                'longitude': geometry['coordinates'][0],
                'depth_km': geometry['coordinates'][2],
                'magnitude': props['mag'],
                'place': props['place'],
                'type': props['type'],
                'status': props['status'],
                'tsunami': props['tsunami'],
                'significance': props['sig'],
                'url': props['url'],
                'load_dt': datetime.now().strftime('%Y-%m-%d')
            }
            earthquakes.append(earthquake)
        
        # Создаём DataFrame
        df = pd.DataFrame(earthquakes)
        
        # Сохраняем в CSV
        df.to_csv('Earthquakes.csv', index=False, encoding='utf-8')
        
        #print(f"Получено {len(df)} землетрясений")
        #print("\nПервые 5 записей:")
        #print(df[['time', 'place', 'magnitude', 'depth_km']].head())
        
        # Статистика
        #print("\nСтатистика по магнитудам:")
        #print(df['magnitude'].describe())
        
        # Самые сильные землетрясения
        #print("\nТоп-5 самых сильных землетрясений:")
        #top5 = df.nlargest(5, 'magnitude')[['time', 'place', 'magnitude', 'depth_km']]
        #print(top5)
        
        return df
        
    except Exception as e:
        print(f"Ошибка: {e}")
        return None

# Запуск
df = get_earthquakes_dataframe()
```
