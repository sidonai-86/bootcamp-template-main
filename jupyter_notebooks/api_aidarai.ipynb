```python
import requests
import pandas as pd
import datetime
import json
import pprint
```

```python
url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
update_at = datetime.datetime.now()
load_date_start = "2025-07-25"
load_date_end = "2026-01-01 "

params = {"format": "geojson", 
          "starttime": load_date_end,
          "minmagnitude": 5.0,
          "orderby": "time"}
```

```python
response = requests.get(url, params=params)
response.raise_for_status()
```

```python
type(response.json()["features"])
```

```python
df = pd.json_normalize(response.json()["features"])

df["event_date"] = pd.to_datetime(df["properties.time"], unit="ms", utc=True)
df["update_at"] = pd.to_datetime(update_at)
df.head()
```

```python
csv_path = f"earthquakes_{update_at}.csv"
df.to_csv(csv_path, index=False)
```

```python
parquet_path = f"earthquakes_{update_at}.parquet"
df.to_csv(parquet_path, index=False)
```
